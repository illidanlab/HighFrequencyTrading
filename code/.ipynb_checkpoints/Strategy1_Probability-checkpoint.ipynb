```python
%%zipline --start 2016-1-1 --end 2018-1-1
## strategy: use a pre-trained model to get probability (or confidence) 
## of your prediction,
## then use this probability to adjust your bet sizing
## %%zipline --start 2010-1-1 --end 2016-1-1

from zipline.api import order_target, record, symbol
import matplotlib.pyplot as plt
import numpy as np
import pickle

def initializeCDF(history, num_bins=20):
    # Use the histogram function to bin the data
    counts, bin_edges = np.histogram(history, bins=num_bins, normed=True) 
    # Now find the cdf
    cdf = np.cumsum(counts) 
    bin_edges=bin_edges[1:]
    return bin_edges, cdf

def getCDF(price, bin_edges, cdf):
    cdfs = [cdf[i] for i, b in enumerate(bin_edges) if price < b ]
    return 1 if len(cdfs) == 0 else cdfs[0] 
    
    
def initialize(context):
    context.i = 0
    context.warmup_i = 1200 
    # we use the first 900 day's data to train the model, and use 300 to test
    context.asset = symbol('AAPL')
    context.model_file = '../model/LR_win7_T5.sav'
    context.model = None
    
    context.history_file = '../data/history.pickle' 
    context.history = None
    
    #context.mode = 'train' 
    context.mode = 'test' 
    #context.mode = 'load' 
    if context.mode == 'test':
        ## read previous model for once
        with open(context.model_file, 'rb') as f:
            context.model = pickle.load(f)  
        
        
            
def handle_data(context, data):   
    context.i += 1
    if context.mode == 'load': 
        ## Skip first warmup_i days 
        if context.i != context.warmup_i:
            return
        history = data.history(context.asset, 'price', bar_count=context.warmup_i, frequency="1d")
        with open(context.history_file, 'wb') as f:
            pickle.dump(history, f)
        return
    
    if context.mode == 'test':
        if context.i <= 10:
            return
        price = data.current(context.asset, 'price')  
        history = data.history(context.asset, 'price', bar_count=6, frequency="1d")
        #print(history)
        history = history.as_matrix()
        window = np.append(history,price)
        
        prob = context.model.predict_proba([window])[0] 
        Ypredict = np.argmax(prob) -1
        p_x = prob[-1] 
        z = (p_x - 1.0/3) / np.sqrt(p_x * (1 - p_x))
        m = 2 * z - 1
        if Ypredict < 0:
            #print('negative')
            order_target(context.asset, 0)
            
        elif Ypredict > 0: 
            print('positive')
            order_target(context.asset, 100) 
         

        # Save values for later inspection
        record( AAPL=data.current(context.asset, 'price'),
               short_pos=price,
               long_pos=price)
  
def analyze(context, perf):
    fig = plt.figure()
    ax2 = fig.add_subplot(111)
    perf['AAPL'].plot(ax=ax2)
    perf[['short_pos', 'long_pos']].plot(ax=ax2)

    perf_trans = perf.ix[[t != [] for t in perf.transactions]]
    buys = perf_trans.ix[
        [t[0]['amount'] > 0 for t in perf_trans.transactions]]
    sells = perf_trans.ix[
        [t[0]['amount'] < 0 for t in perf_trans.transactions]]
    
    ax2.plot(buys.index, perf.long_pos.ix[buys.index],
             '^', markersize=10, color='m')
    ax2.plot(sells.index, perf.short_pos.ix[sells.index],
             'v', markersize=10, color='k')
    ax2.set_ylabel('price in $')
    plt.legend(loc=0) 
    plt.show() 

 
    
```
